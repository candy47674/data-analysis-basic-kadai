```python
from tensorflow.keras.applications.vgg16 import VGG16 #VGG16モデルインストール
model = VGG16(weights='imagenet')
model.summary() #importしたVGG16の学習データを表で可視化確認
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
#画像をプロセッシング(前処理)するためのプログラムをインポート
# load_img画像サイズ変更プログラム,img_to_array画像配の列変換プログラム

unknown_img = load_img('/content/unknown_vgg.jpg', target_size=(224,224))
# 画像をインポートしてサイズを指定

unknown_img
import matplotlib.pyplot as plt
# 画像を表示
plt.imshow(unknown_img)
plt.axis('off')  # XY軸を表示しない
plt.show()

# img_to_arrayは数値に変換するプログラム
unknown_array = img_to_array(unknown_img)
# unknown_array.shape　224,224,3

unknown_array = unknown_array.reshape(1,224,224,3)
# 1画像　224，224画像の幅と高さ3←3次元
unknown_array.shape

from tensorflow.keras.applications.vgg16 import preprocess_input
unknown_array = preprocess_input(unknown_array)
# ピクセル値のスケーリング（正規化）。
# モデルに合わせたピクセル値の調整（モデルごとに異なる処理が必要）。
result = model.predict(unknown_array)
result
from tensorflow.keras.applications.vgg16 import decode_predictions
decode_predictions(result)
```

```python
from tensorflow.keras import utils
utils.set_random_seed(0) #乱数固定
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# 画像を前処理するための処理のモデルをインストール

datagen = ImageDataGenerator(rescale=1/255,validation_split=0.25)
#dog,catデータをインスタンス化（正規化）
train_generator = datagen.flow_from_directory(
    '/content/drive/MyDrive/data_train',
    subset='training',
    target_size=(256,256),
    batch_size=64
)
train_generator.class_indices
# cat:0 dog:1と表示確認
val_generator = datagen.flow_from_directory(
    '/content/drive/MyDrive/data_train',
    subset = 'validation',
    target_size=(256,256),
    batch_size=64
)
test_generator = datagen.flow_from_directory(
    '/content/drive/MyDrive/data_test',
    target_size =(256,256),
    batch_size =64
)
from tensorflow.keras.applications.vgg16 import VGG16
vgg16_without_fc = VGG16(weights='imagenet',include_top=False,input_shape=(256,256,3))
vgg16_without_fc.summary()
from tensorflow.keras.models import Sequential
model = Sequential()
for layer in vgg16_without_fc.layers:
  model.add(layer)
model.summary()
for layer in model.layers:
  layer.trainable = False
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
# 3つのモデルをインポート

# Sequentialモデルの定義
model = Sequential()

model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(2,activation='softmax'))
model.summary
```

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)

```python
import pandas as pd, seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame(history.history)
sns.lineplot(data=df[['loss','val_loss']])
sns.lineplot()
```
